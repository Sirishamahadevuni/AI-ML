append() Function

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Chemistry':[98,48,90,49,70], 'Physics':[98,48,90,49,70]})
df2=pd.DataFrame({'Roll No':[6,7,8,9,10], 'Chemistry':[78,98,98,89,72],'Physics':[98,48,90,49,70]})
print(df1)
print(df2)
```

```python
df1._append(df2)
```

```python
df1._append(df2,ignore_index=True)
```

```python
df1._append(df2,ignore_index=True,sort=True)
```
