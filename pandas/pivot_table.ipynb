# pivot()
pivot allows to transform or reshape data

```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\siris\OneDrive\Desktop\AIML\pandas\Datasets\sample2.csv")
df.head()
```

```python
df['DOB']=pd.to_datetime(df['DOB'],format='%d-%m-%Y')
```

```python
pivot=df.pivot_table(index='Branch',values=['Physics','Chemistry','Maths','Computer'],aggfunc='mean')
```

```python
pivot
```

```python
df.pivot_table(index='Branch',columns='Section',values=['Physics','Chemistry','Maths'],aggfunc='mean')
```
