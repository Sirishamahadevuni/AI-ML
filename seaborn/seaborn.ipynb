```python
# Scatter Plot
```

```python
titanic_df=sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df)
```

```python
sns.scatterplot(x='age',y='fare',
                data=titanic_df,hue='alive')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',
                data=titanic_df,hue='alive',
                style='class')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',
                data=titanic_df,hue='alive',
                style='class',
                palette='inferno')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',
                data=titanic_df,hue='alive',
                style='class',
                palette='gist_rainbow',alpha=0.8)
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='age',y='fare',
                data=titanic_df,hue='alive',
                style='class',
                palette='gist_rainbow',alpha=0.8)
sns.lineplot(x='age',y='fare',data=titanic_df,color='green')
plt.title('Titanic Data Analysis')
```

Bar Plots

```python
titanic_df.head()
```

```python
sns.barplot(x='class',y='fare',data=titanic_df)
```

bar plot represents an estimate tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainity around that estimate using error bars

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex')
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex',palette='inferno')
```

```python
sns.barplot(y='class',x='fare',data=titanic_df,
            hue='sex',palette='inferno',
            orient='h')
```

```python
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex',
            palette='inferno',
            estimator=np.max)
```

```python
# Confidence Interval
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex',
            palette='inferno',
            ci=20,
            errcolor='red',
            errwidth=8)
```

```python
#saturation
sns.barplot(x='class',y='fare',data=titanic_df,
            hue='sex',
            palette='inferno',
            saturation=0.3)
```
