```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
flight_df = pd.read_csv(r'C:\Users\siris\OneDrive\Desktop\AIML\seaborn\flights.csv')
flight_df.head()
```

```python
pivot_table = flight_df.pivot_table(index='month', columns='year', values='passengers', aggfunc='sum')
```

```python
pivot_table.isnull().sum().sum()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='Blues', annot=True, fmt='g')
plt.title('Heatmap of Flight Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```
